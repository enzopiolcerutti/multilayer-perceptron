# Experimentos 

Este notebook mostra os experimentos da atividade. A implementação principal está em `mlp/`, usa NumPy para forward pass, backpropagation e SGD.

## Ideia tecnica

A rede recebe imagens 28 x 28 achatadas para vetores de 784 entradas. As camadas ocultas usam Relu, e a camada de saída usa softmax. A loss e cross-entropy. No backward, a derivada da combinação softmax + cross-entropy simplifica para `(y_pred - y_true) / batch_size`.

````python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from mlp import MLP
from mlp.losses import one_hot
import numpy as np
import pandas as pd
````

## Sanity check com XOR

O XOR é pequeno, mas não é linearmente separável. Se a MLP aprende XOR, ocorre a primeira evidência de que as camadas ocultas, ativação não linear e backprop estão funcionando.

````python
x = np.array([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
y = one_hot(np.array([0, 1, 1, 0]), 2)

model = MLP([2, 8, 8, 2], learning_rate=0.2, seed=10)
history = model.fit(x, y, epochs=2000, batch_size=4, x_val=x, y_val=y, verbose=False)
model.predict(x)
````
